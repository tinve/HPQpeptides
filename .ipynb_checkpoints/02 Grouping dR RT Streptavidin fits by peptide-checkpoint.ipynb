```python
from __future__ import division

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats

% matplotlib inline
```

# Helper function, finds weight of aminoacid

```python
def weight(name, dictionary):
    w = 0
    for e in name:
        w = w + dictionary[e]
    return int(w)
```

# Aminoacid weights to dictionary

```python
amino = pd.read_csv('Aminoacids.csv', header = 0)
amino.head()
```

```python
amino.drop('Aminoacid', axis = 1, inplace = True)
amino.columns = ['Code', 'Weight']
weights = amino.set_index('Code')['Weight'].to_dict()
weights
```

# Peptide data cleaning

```python
df = pd.read_csv('01 Filtered dR Streptavidin fit.csv', index_col = 'Unnamed: 0')

keep_columns = ['Peptide',
                'Reaction',
                'Kd filtered',
                'Ka filtered']
df = df[keep_columns]

# remove NaNs and Xs
df = df.dropna()
df = df[ df['Peptide'] != 'X' ]

df.sort('Peptide', inplace = True)
df.reset_index(drop = True, inplace = True)
df.head()
```

# Calculate mean and standard deviation by peptide

```python
f = {'Kd filtered' : [len, np.mean, np.std, stats.sem],
     'Ka filtered' : [np.mean, np.std, stats.sem],
     'Reaction'    : [sum]}

result = df.groupby(['Peptide']).agg(f)

result.reset_index(inplace = True)
result.columns = result.columns.droplevel()

result.columns = ['Peptide',
                  'Number of Spots',
                  
                  'Kd Filtered Mean',
                  'Kd Filtered SD',
                  'Kd Filtered SE',
                  
                  'Ka Filtered Mean',
                  'Ka Filtered SD',
                  'Ka Filtered SE',

                  'Number of Reactions']

result['Reaction Fraction'] = result['Number of Reactions'] / result['Number of Spots']

result.fillna(value = 0, inplace = True)

# explicitly set SD and SE to zero when 0% reactions
result['Kd Filtered SD'] = result['Kd Filtered SD'] * (result['Number of Reactions'] != 0).astype(int)
result['Ka Filtered SD'] = result['Ka Filtered SD'] * (result['Number of Reactions'] != 0).astype(int)
result['Kd Filtered SE'] = result['Kd Filtered SE'] * (result['Number of Reactions'] != 0).astype(int)
result['Ka Filtered SE'] = result['Ka Filtered SE'] * (result['Number of Reactions'] != 0).astype(int)

result['Peptide Length'] = result['Peptide'].map(len)
result['Peptide Weight'] = result['Peptide'].map(lambda x: weight(x, weights))
result.sort(['Peptide Length', 'Peptide'], inplace = True)
result.reset_index(drop = True, inplace = True)

result = result[['Peptide',
                 'Peptide Length',
                 'Peptide Weight',
                 'Number of Spots',
                 'Number of Reactions',
                 'Reaction Fraction',
                 'Kd Filtered Mean',
                 'Kd Filtered SD',
                 'Kd Filtered SE',
                 'Ka Filtered Mean',
                 'Ka Filtered SD',
                 'Ka Filtered SE',]]

result.head()
```

```python
peptides_mixed = result[~result['Reaction Fraction'].isin([1, 0])]['Peptide']
peptides_single = result[(result['Number of Spots'] == 1) & (result['Reaction Fraction'] == 1)]['Peptide']

df = pd.read_csv('01 Filtered dR Streptavidin fit.csv', index_col = 'Unnamed: 0')

indices_mixed = []
indices_single = []

for i in range(3):
    indices = (df[(df['Peptide'].isin(peptides_mixed)) & (df['Chunk'] == i)]['Set']).tolist()
    indices_mixed = indices_mixed + [len(indices)] + indices
    
    indices = (df[(df['Peptide'].isin(peptides_single)) & (df['Chunk'] == i)]['Set']).tolist()
    indices_single = indices_single + [len(indices)] + indices    

f = open('set indices mixed response.txt', 'w')
f.write('\n'.join([str(e) for e in indices_mixed]))
f.close()
    
f = open('set indices single reaction.txt', 'w')
f.write('\n'.join([str(e) for e in indices_single]))
f.close()
```
